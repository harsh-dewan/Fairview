```python
import pandas as pd
from pymongo import MongoClient
import matplotlib.pyplot as plt
import json
import seaborn as sns
import numpy as np
```

```python
client=MongoClient("localhost",27017)
```

```python
db = client["educationDB"]
collection = db["educationLiteracy"]
```

```python
# Retrieve all documents from events collection
documents = collection.find().sort([("Year")])
# Iterate over the documents
data = list(documents)
#print(data)
mongo_df = pd.DataFrame(data)
x = mongo_df["Year"]
y = mongo_df["World, Female"]
z = mongo_df["World, Male"]
t = mongo_df["World, Total"]
plt.rcParams["figure.figsize"] = [10,5]
plt.rcParams["figure.autolayout"] = True
plt.plot(x, y, label="Literacy rate Females") 
plt.plot(x, z, label="Literacy rate Males") 
plt.plot(x, t, label="Literacy rate Total")
dimX = np.arange(mongo_df["Year"].min(),mongo_df["Year"].max(),5)
dimY = np.arange(0,100,10)
plt.xticks(dimX)
plt.yticks(dimY)
plt.legend()

# function to show the plot 
plt.show()
#for document in documents:
    #print(document)
```
